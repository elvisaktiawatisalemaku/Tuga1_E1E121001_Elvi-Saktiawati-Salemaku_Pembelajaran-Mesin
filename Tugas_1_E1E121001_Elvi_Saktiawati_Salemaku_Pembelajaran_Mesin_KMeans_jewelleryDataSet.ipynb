```python
df = df.rename(columns={'Age':'age','Income':'income', 'SpendingScore':'spending score', 'Savings':'savings' })
```

```python
X=df
```

```python
X.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Menghitung nilai inersia untuk jumlah cluster 1 hingga 15
inertias = []
for i in range(1, 16):
    km = KMeans(n_clusters=i)
    km.fit(X)
    inertias.append(km.inertia_)

```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns


# Plot nilai inersia

plt.plot(range(1, 16), inertias, marker='o')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Inersia')
plt.title('Elbow Method untuk Menentukan Jumlah Cluster Optimal')

```
